## Teoría del Confort Térmico en Ganado Vacuno

El confort térmico en el ganado vacuno se refiere al rango de condiciones ambientales donde los animales no necesitan gastar energía adicional para mantener su temperatura corporal normal. Cuando las condiciones ambientales (temperatura, humedad, radiación solar y velocidad del viento) exceden o caen por debajo de este rango, el ganado experimenta estrés térmico, lo que puede afectar negativamente su salud, bienestar y productividad.

**Factores que influyen en el confort térmico:**

*   **Temperatura del Aire:** Es el factor más directo que afecta la transferencia de calor entre el animal y su entorno.
*   **Humedad Relativa:** Una alta humedad reduce la capacidad del animal para disipar calor a través de la evaporación (sudoración y jadeo).
*   **Radiación Solar:** La exposición directa al sol aumenta la carga de calor en los animales.
*   **Velocidad del Viento:** Una brisa puede ayudar a disipar el calor corporal, pero vientos fuertes en temperaturas bajas pueden aumentar la pérdida de calor.
*   **Raza y Etapa Productiva:** Algunas razas (ej. razas índicas) son más tolerantes al calor que otras (ej. razas taurinas). Animales en producción (ej. vacas lecheras de alta producción) son más susceptibles al estrés por calor debido a su alta tasa metabólica.

**Efectos del Estrés Térmico:**

El estrés térmico puede llevar a una serie de problemas, incluyendo:

*   **Disminución del Consumo de Alimento:** Los animales reducen su ingesta para disminuir la producción de calor metabólico.
*   **Reducción de la Producción:** Menor producción de leche, carne y menor eficiencia reproductiva.
*   **Cambios Fisiológicos:** Aumento de la frecuencia respiratoria (jadeo), aumento de la temperatura corporal, cambios hormonales.
*   **Aumento de la Susceptibilidad a Enfermedades:** El estrés compromete el sistema inmunológico.
*   **Mortalidad:** En casos extremos, el estrés térmico severo puede ser fatal.

**Índices de Confort Térmico:**

Para cuantificar el riesgo de estrés térmico, se utilizan índices que combinan múltiples variables ambientales en un solo valor. El Índice de Temperatura y Humedad (ITH) es uno de los más comunes y se utiliza para evaluar el nivel de confort o estrés térmico en el ganado, ayudando a los productores a tomar decisiones informadas para mitigar sus efectos.

# **Métodos y cálculos**

+ **Índice de temperatura y humedad (producción intensiva)**

$$𝐼𝑇𝐻 = 0.72(𝑇_{𝑏𝑠} + 𝑇_{𝑏ℎ}
) + 40.6 [1]$$

Donde:

1. $T_{bs}$: Temperatura bulbo seco ($^oC$)

2. $T_{bh}$: Temperatura de bulbo húmedo ($^oC$)

Para la correlación de los datos de la especie bovina en la producción de leche se recomienda utilizar los límites propuestos por National Weather Service:

1. Sí $ITH <= 70$ ambiente no estresante

2. Sí $ITH > 70 \space and \space ITH <= 78$ situación de alerta por estrés térmico

3. Sí $ITH > 78 \space and \space ITH <= 83$ peligro por estrés térmico

4. Sí $ITH > 83$ situación crítica


````python
# Entrada: pedir tbs (Temperatura de Bulbo seco) y tbh (Temperatura de Bulbo humedo)
tbs = float(input("Ingrese la temperatura de bulbo seco, Grados centigrados: "))
tbh = float(input("Ingrese la temperatura de bulbo húmedo, Grados centigrados: "))

# Proceso: Calculo de ith (indice de Temperatura de humedad)
ith = 0.72 * (tbs + tbh) + 40.6

# Salida: Mostrar el resultado del calculo, de la evalucion del ith y el confort térmico
if ith <= 70:
    print("El ambiente no es estresante")
elif ith > 70 and ith <= 78:
    print("La situación de alerta por estrés térmico")
elif ith > 78 and ith <= 83:
    print("peligro por estrés térmico")
elif ith > 83:
    print("situación crítica")
````
