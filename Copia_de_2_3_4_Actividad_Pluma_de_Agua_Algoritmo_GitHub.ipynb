## 6. Espacio para que el estudiante construya la solución

### Instrucciones
Complete el siguiente bloque de código a partir del algoritmo anterior.

````python
# ==========================================================
# ACTIVIDAD: MODELAR LA "PLUMA DE AGUA" CON PYTHON
# ==========================================================

# 1. Solicitar al usuario el número de plumas
# Escriba aquí su código
plumas = float(input("Ingrese el número de plumas: "))

# 2. Definir constantes del problema
# Escriba aquí su código
caudal_pluma = 0.025
segundos_por_dia = 86400


# 3. Calcular el caudal total
# Escriba aquí su código

caudal_total = (plumas * caudal_pluma)

# 4. Calcular el consumo diario
# Escriba aquí su código
consumo_diario = (caudal_total * segundos_por_dia)

# 5. Calcular el consumo mensual
# Escriba aquí su código
consumo_mensual = (consumo_diario * 30)

# 6. Clasificar el consumo
# Escriba aquí su código
if consumo_diario < 1000:
    nivel_consumo = "Bajo"
elif consumo_diario >= 1000 and consumo_diario <= 3000:
    nivel_consumo = "Medio"
else:
    nivel_consumo = "Alto"

# 7. Mostrar resultados
# Escriba aquí su código
print(f"Caudal total: {caudal_total:.3f} litros por segundo")
print(f"Consumo diario: {consumo_diario:.3f} litros")
print(f"Consumo mensual: {consumo_mensual:.3f} litros")
print(f"Clasificación del consumo: {nivel_consumo}")
````

## 8. Extensión opcional

Si desea profundizar, puede ampliar el programa para que también permita:

````python
# ==========================================================
# ACTIVIDAD: MODELAR LA "PLUMA DE AGUA" CON PYTHON EXTENDIDO
# ==========================================================

# 1. Solicitar al usuario el número de plumas
# Se solicita al usuario que ingrese la cantidad de plumas de agua.
# Se utiliza float() para permitir valores decimales.
plumas = float(input("Ingrese el número de plumas: "))

# 2. Definir constantes del problema y nuevas constantes para extensiones
# Constante: caudal de una pluma de agua en litros por segundo.
CAUDAL_PLUMA = 0.025
# Constante: número de segundos en un día (24 horas * 60 minutos * 60 segundos).
SEGUNDOS_POR_DIA = 86400
# Constante: tarifa por litro de agua para calcular el costo (ejemplo en COP).
TARIFA_POR_LITRO = 0.002
# Constante: factor de conversión de litros a metros cúbicos (1 m³ = 1000 litros).
LITROS_POR_M3 = 1000

# 3. Calcular el caudal total
# El caudal total se calcula multiplicando el número de plumas por el caudal de una pluma.
caudal_total = (plumas * CAUDAL_PLUMA)

# 4. Calcular el consumo diario
# El consumo diario en litros se obtiene multiplicando el caudal total por los segundos en un día.
consumo_diario = (caudal_total * SEGUNDOS_POR_DIA)

# 5. Calcular el consumo semanal y mensual
# El consumo semanal se calcula multiplicando el consumo diario por 7 días.
consumo_semanal = (consumo_diario * 7)
# El consumo mensual se calcula multiplicando el consumo diario por 30 días (valor aproximado).
consumo_mensual = (consumo_diario * 30)

# 6. Clasificar el consumo
# Se utiliza una estructura condicional if-elif-else para clasificar el consumo diario
# en 'Bajo', 'Medio' o 'Alto' según los rangos definidos.
if consumo_diario < 1000:
    nivel_consumo = "Bajo"
elif 1000 <= consumo_diario <= 3000:
    nivel_consumo = "Medio"
else:
    nivel_consumo = "Alto"

# 7. Calcular costo y convertir a m3 (Extensiones)
# El costo mensual se estima multiplicando el consumo mensual por la tarifa por litro.
costo_mensual = consumo_mensual * TARIFA_POR_LITRO
# El consumo mensual se convierte a metros cúbicos dividiéndolo por el factor de litros por m³.
consumo_mensual_m3 = consumo_mensual / LITROS_POR_M3

# 8. Mostrar resultados (actualizado con extensiones)
# Se imprimen todos los resultados calculados, formateando los números a 3 decimales
# y el costo mensual a 2 decimales, e incluyendo las unidades correspondientes.
print(f"Caudal total: {caudal_total:.3f} litros por segundo")
print(f"Consumo diario: {consumo_diario:.3f} litros")
print(f"Consumo semanal: {consumo_semanal:.3f} litros")
print(f"Consumo mensual: {consumo_mensual:.3f} litros ({consumo_mensual_m3:.3f} m³)")
print(f"Costo mensual estimado: {costo_mensual:.2f} COP")
print(f"Clasificación del consumo: {nivel_consumo}")
````

### 6. Versión Comentada del Código

```python
# ==========================================================
# ACTIVIDAD: MODELAR LA "PLUMA DE AGUA" CON PYTHON EXTENDIDO
# ==========================================================

# 1. SOLICITUD DE ENTRADA AL USUARIO
# Se solicita al usuario que ingrese la cantidad de plumas de agua.
# Se utiliza float() para permitir valores decimales y mayor precisión.
plumas = float(input("Ingrese el número de plumas: "))

# 2. DEFINICIÓN DE CONSTANTES DEL PROBLEMA
# Nombres de constantes en mayúsculas (CAUDAL_PLUMA) para buenas prácticas.

CAUDAL_PLUMA = 0.025         # Litros por segundo que aporta una pluma.
SEGUNDOS_POR_DIA = 86400     # Segundos en un día (24 horas * 60 minutos * 60 segundos).
TARIFA_POR_LITRO = 0.002     # Tarifa de costo por litro de agua (ejemplo en COP).
LITROS_POR_M3 = 1000         # Factor de conversión de litros a metros cúbicos (1 m³ = 1000 L).

# 3. CÁLCULOS PRINCIPALES

# Caudal total: Multiplica el número de plumas por el caudal de una pluma.
caudal_total = plumas * CAUDAL_PLUMA

# Consumo diario: Caudal total multiplicado por los segundos en un día.
consumo_diario = caudal_total * SEGUNDOS_POR_DIA

# Consumo semanal: Consumo diario multiplicado por 7 días.
consumo_semanal = consumo_diario * 7

# Consumo mensual: Consumo diario multiplicado por 30 días (aproximación).
consumo_mensual = consumo_diario * 30

# 4. CLASIFICACIÓN DEL CONSUMO
# Uso de if-elif-else para categorizar el consumo diario.

if consumo_diario < 1000:
    nivel_consumo = "Bajo"
elif 1000 <= consumo_diario <= 3000:
    nivel_consumo = "Medio"
else:
    nivel_consumo = "Alto"

# 5. EXTENSIONES ADICIONALES (Costo y Conversión a m³)

# Costo mensual: Consumo mensual en litros por la tarifa por litro.
costo_mensual = consumo_mensual * TARIFA_POR_LITRO

# Consumo mensual en m³: Consumo mensual en litros dividido por LITROS_POR_M3.
consumo_mensual_m3 = consumo_mensual / LITROS_POR_M3

# 6. PRESENTACIÓN DE RESULTADOS
# Impresión de todos los resultados, formateando números a 3 decimales (o 2 para costo).

print(f"Caudal total: {caudal_total:.3f} litros por segundo")
print(f"Consumo diario: {consumo_diario:.3f} litros")
print(f"Consumo semanal: {consumo_semanal:.3f} litros")
print(f"Consumo mensual: {consumo_mensual:.3f} litros ({consumo_mensual_m3:.3f} m³)")
print(f"Costo mensual estimado: {costo_mensual:.2f} COP")
print(f"Clasificación del consumo: {nivel_consumo}")
```
