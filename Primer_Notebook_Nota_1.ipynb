# Problema de Ingeniería (Nivel Básico)

**Cálculo del volumen de agua necesario para llenar un tanque rectangular.**

Este tipo de cálculo es común en **ingeniería civil, agrícola o agroindustrial**, por ejemplo para diseñar **reservorios de agua o tanques de almacenamiento**.

La fórmula del volumen de un prisma rectangular es:

[
V = largo \times ancho \times altura
]

Datos que el usuario debe proporcionar:

* `altura` del tanque (m)

---

### Proceso

1. Leer el largo.

[
volumen = largo \times ancho \times altura

### Ejemplo de Lógica Programática

Para mostrarte cómo funciona, pensemos en un problema sencillo de **Ingeniería Agrícola**: calcular el volumen de agua necesario para un reservorio cilíndrico.

Antes de darte el código "mágico", analicemos la lógica. La fórmula del volumen de un cilindro es:

$$V = \pi \cdot r^2 \cdot h$$

Donde:

* $r$ es el **radio** del reservorio.
* $h$ es la **altura** (o profundidad).

#### Propuesta de Código (Estilo Profesional)

Aquí tienes un ejemplo de cómo estructuraríamos esto en Python, usando nombres de variables descriptivos y una función modular:

````python
import math

def calcular_volumen_reservorio(radio_m, altura_m):
    """
    Calcula el volumen de un reservorio cilíndrico en metros cúbicos.
    """
    volumen = math.pi * (radio_m ** 2) * altura_m
    return volumen

# Datos de entrada (Imagina que vienen de un sensor o un formulario de Streamlit)
radio_proyecto = 2.5  # metros
altura_proyecto = 4.0  # metros

# Llamada a la función
resultado_v = calcular_volumen_reservorio(radio_proyecto, altura_proyecto)

print(f"El volumen total del reservorio es: {resultado_v:.2f} m3")
````
