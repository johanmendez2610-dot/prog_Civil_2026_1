# Problema de Ingeniería (Nivel Básico)

**Cálculo del volumen de agua necesario para llenar un tanque rectangular.**

Este tipo de cálculo es común en **ingeniería civil, agrícola o agroindustrial**, por ejemplo para diseñar **reservorios de agua o tanques de almacenamiento**.

La fórmula del volumen de un prisma rectangular es:

[
V = largo \times ancho \times altura
]

---

# 1. Definición del Algoritmo

Primero pensamos el problema **antes de programar**.

### Entradas (Inputs)

Datos que el usuario debe proporcionar:

* `largo` del tanque (m)
* `ancho` del tanque (m)
* `altura` del tanque (m)

---

### Proceso

1. Leer el largo.
2. Leer el ancho.
3. Leer la altura.
4. Calcular el volumen:

[
volumen = largo \times ancho \times altura
]

---

### Salidas (Outputs)

* Mostrar el **volumen del tanque en metros cúbicos (m³)**.

---

# 2. Pseudocódigo

Antes de escribir Python, usamos pseudocódigo.

```
INICIO

leer largo
leer ancho
leer altura

volumen = largo * ancho * altura

mostrar volumen

FIN
```

---

# 3. Código en Python

```python
# -----------------------------------------
# Programa: Cálculo del volumen de un tanque
# Autor: Estudiante de Ingeniería
# -----------------------------------------

# Entradas
largo = float(input("Ingrese el largo del tanque (m): "))
ancho = float(input("Ingrese el ancho del tanque (m): "))
altura = float(input("Ingrese la altura del tanque (m): "))

# Proceso
volumen = largo * ancho * altura

# Salida
print("El volumen del tanque es:", volumen, "m^3")
```

---

# 4. Ejemplo de Ejecución

Entrada del usuario:

```
largo = 5
ancho = 3
altura = 2
```

Cálculo:

```
volumen = 5 * 3 * 2
volumen = 30
```

Salida:

```
El volumen del tanque es: 30 m^3
```
